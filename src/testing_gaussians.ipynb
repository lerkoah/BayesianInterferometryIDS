```python
import numpy as np
import theano.tensor as tt
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0,2*np.pi,1000)
f = np.sin(10*t)


plt.plot(t,f)

g = np.exp(-0.5*((t-np.pi)**2)/(1))
plt.plot(t,g)
plt.plot(t,g*f)
plt.show()
```

```python
numberOfSamplings = 1000
frecuency = np.pi*5

x = np.linspace(0,2*np.pi,numberOfSamplings)
y = np.linspace(0,2*np.pi,numberOfSamplings)

X, Y = np.meshgrid(x,y)

z1 = np.zeros((numberOfSamplings,numberOfSamplings))
z2 = np.zeros((numberOfSamplings,numberOfSamplings))

for i in range(numberOfSamplings):
    for j in range(numberOfSamplings):
        z1[i,j] = np.sin(frecuency*x[i]+frecuency*y[j])
        z2[i,j] = np.sin(frecuency*x[i]+frecuency*y[j])
```

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm


fig = plt.figure(figsize=plt.figaspect(0.5))
ax = fig.add_subplot(2, 1, 1, projection='3d')
surf = ax.plot_surface(X,Y, z1, cmap=cm.viridis)
fig.colorbar(surf, shrink=0.5, aspect=5)

ax = fig.add_subplot(2, 1, 2, projection='3d')
surf = ax.plot_surface(X,Y, z2, cmap=cm.viridis)
fig.colorbar(surf, shrink=0.5, aspect=5)

ax.view_init(elev=20., azim = 45)
fig.set_figheight(20)
fig.set_figwidth(26)
plt.show()
```

```python
gg1 = np.zeros((numberOfSamplings,numberOfSamplings))
gg2 = np.zeros((numberOfSamplings,numberOfSamplings))

for i in range(numberOfSamplings):
    for j in range(numberOfSamplings):
        gg1[i,j] = np.exp(-0.5*((x[i]-np.pi)**2)/(0.1))
        gg2[i,j] = np.exp(-0.5*((y[j]-np.pi)**2)/(0.1))

        
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm


fig = plt.figure(figsize=plt.figaspect(0.5))
ax = fig.add_subplot(2, 1, 1, projection='3d')
surf = ax.plot_surface(X,Y, gg1, cmap=cm.viridis)
fig.colorbar(surf, shrink=0.5, aspect=5)

ax = fig.add_subplot(2, 1, 2, projection='3d')
surf = ax.plot_surface(X,Y, gg2, cmap=cm.viridis)
fig.colorbar(surf, shrink=0.5, aspect=5)
#ax.view_init(elev=10., azim = 70)
fig.set_figheight(20)
fig.set_figwidth(26)
plt.show()
```

```python
fig = plt.figure(figsize=plt.figaspect(0.5))
ax = fig.add_subplot(1, 1, 1, projection='3d')
surf = ax.plot_surface(X,Y, np.log(gg1*gg2), cmap=cm.viridis)
fig.colorbar(surf, shrink=0.5, aspect=5)

#ax.view_init(elev=10., azim = 45)
fig.set_figheight(20)
fig.set_figwidth(26)
plt.show()
```
